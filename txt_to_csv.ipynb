```python
dfup = pd.read_excel('trans_update.xlsx')

dfup.fillna(method='bfill', inplace=True)

dfup
```

```python
# 降维处理
dfup.drop(['date1', 'date2'], axis=1, inplace=True)
# 索引重构
#trans_newIndex = {'transcript': 'Trans', 'date_3': 'Date'}
#dfup.rename(columns=trans_newIndex, inplace=True)
dfup


```

```python
# 预防报错：None 错误-ValueError: (u'Unknown string format:', 'None')
#dfup['Date'] = dfup['Date'].replace({'None': np.NaN})

# 'float' to 'String' type. 
dfup['Date'] = dfup['Date'].astype(str)
dfup
```

```python
dfup.info
```

```python
# 时间序列处理
# 因时间日期格式非统一，故选用dateutil包对parser.parse方法识别多变时间格式
dfup['Date'] = dfup['Date'].apply(lambda dt: dt_parse(dt))

# 按date升序排列
dfup.sort_values(by='Date', inplace=True, ascending=True)

dfup
```

```python
dfup.to_csv('TSLA_transcript.csv',encoding="utf_8_sig", index=False)
```
