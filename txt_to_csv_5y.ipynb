```python
# 预防报错：None 错误-ValueError: (u'Unknown string format:', 'None')
#dfup['Date'] = dfup['Date'].replace({'None': np.NaN})

# 'float' to 'String' type. 
dfup['Date'] = dfup['Date'].astype(str)
dfup
```

```python
# 时间序列处理
# 因时间日期格式非统一，故选用dateutil包对parser.parse方法识别多变时间格式
dfup['Date'] = dfup['Date'].apply(lambda dt: dt_parse(dt))

# 按date升序排列
dfup.sort_values(by='Date', inplace=True, ascending=True)

dfup
```
